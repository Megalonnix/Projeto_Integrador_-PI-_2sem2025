```python
from openpyxl.utils import get_column_letter
```

```python
primeira_coluna_idx = 3 # Coluna C
ultima_coluna_idx = sheet.max_column # Última Coluna df tabela278.xlsx

blocos_colunas = []

for i in range(primeira_coluna_idx, ultima_coluna_idx + 1, 3):
    bloco_de_tres = (
        get_column_letter(i), 
        get_column_letter(i + 1), 
        get_column_letter(i + 2)) # (total, governo, privado)
    blocos_colunas.append(bloco_de_tres)

pd.DataFrame(blocos_colunas)
```

```python
# Lista para armazenar todos os blocos processados
todos_blocos = []
tipos_armazens = [
    'Armazéns convencionais, estruturais e infláveis',
    'Armazéns graneleiros e granelizados',
    'Silos'
]

ano_inicial = 2007
semestre_inicial = 2  # começa no 2º semestre 2007

for idx, (col1, col2, col3) in enumerate(blocos_colunas):
    # Reorganiza cada coluna
    armazens_totais = reorganizeColumn(col1, cabecalho=tp_armazens_graneleiros[0:6])
    armazens_governo = reorganizeColumn(col2, cabecalho=tp_armazens_graneleiros[6:12])
    armazens_privados = reorganizeColumn(col3, cabecalho=tp_armazens_graneleiros[12:18])
    
    # Tipo de Armazem muda ciclicamente entre os 3 tipos:
    tipo_atual = tipos_armazens[idx % 3]

    # Calcular semestre e ano
    semestre_offset = idx // 3  # quantos semestres já passaram
    ano = ano_inicial + (semestre_inicial - 1 + semestre_offset) // 2
    semestre = (semestre_inicial - 1 + semestre_offset) % 2 + 1
    data_label = f"{semestre}º semestre {ano}"

    # Merge dos 3 blocos
    df_semestre = merge_total_governo_privado_tpArmazem_dt_registro(
        armazens_totais,
        armazens_governo,
        armazens_privados,
        'categoria_armazem',
        tipo_atual,
        'dt_registro',
        data_label,
        novo_df  # usado apenas para definir tamanho
    )
    # Adiciona à lista
    todos_blocos.append(df_semestre)

# Concatena verticalmente todos os semestres
df_final = pd.concat(todos_blocos, axis=0).reset_index(drop=True)
df_final.info()
```
